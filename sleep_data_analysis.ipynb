```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ["3-5", "6-8", "9-11", "12-14", "15-18"]
x_placeholders = np.arange(len(x))
y_good = [11.5, 10, 10, 10, 9]
y_actual = [9.68, 8.98, 8.85, 8.05, 7.40]
width = 0.4

plt.style.use("ggplot")
plt.title("Amount of sleep by Age")
plt.xlabel("Age range (years)")
plt.ylabel("Amount of sleep (hours)")

plt.bar(x_placeholders - 1/2 * width, y_good, width=width, label="Needed")
plt.bar(x_placeholders + 1/2 * width, y_actual, width=width, label="Actual")
plt.xticks(x_placeholders, x)
plt.legend()

plt.savefig('adolescent.png', bbox_inches='tight')
```
